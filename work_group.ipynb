```python
#导入库

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
```

#### pyplot方式绘图

```python
#方式一：pyplot隐式创建和管理Figure、Axes

plt.figure(figsize=(7, 4), layout='constrained')

#绘制曲线图
x = np.linspace(0, 2, 100)
plt.subplot(1, 3, 1)
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot([0.2, 0.25, 1.0, 1.5], [1, 1.5, 2, 3.5])
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()

#绘制散点图
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.subplot(1, 3, 2)
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')

#绘制柱状图
names = ['a', 'b', 'c']
values = [10, 20, 30]

plt.subplot(1, 3, 3)
plt.bar(names, values)
#plt.barh绘制横向条形图
#plt.barh(names, values)

plt.show()
```

#### OO-style绘图
##### 直接调用Matplotlib API修改Figure或Axes

```python
#绘制图和轴
fig, ax = plt.subplots()

#设置轴范围
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

#设置标题和标签
ax.set_title(' a Figure - Matplotlib OO-style Demo', fontsize=14, pad=15)
ax.set_xlabel('x Axis label', fontsize=12)
ax.set_ylabel('y Axis label', fontsize=12)

#设置主要刻度
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(1))
def x_major_formatter(x, pos):
    return f'{x:.0f}'
ax.xaxis.set_major_formatter(FuncFormatter(x_major_formatter))
def y_major_formatter(y, pos):
    return f'{y:.0f}'
ax.yaxis.set_major_formatter(FuncFormatter(y_major_formatter))

#设置次要刻度及其标签
ax.xaxis.set_minor_locator(MultipleLocator(0.25))
ax.yaxis.set_minor_locator(MultipleLocator(0.25))

def x_minor_formatter(x, pos):
    return f'{x:.2f}'
ax.xaxis.set_minor_formatter(FuncFormatter(x_minor_formatter))
def y_minor_formatter(y, pos):
    return f'{y:.2f}'
ax.yaxis.set_minor_formatter(FuncFormatter(y_minor_formatter))

ax.tick_params(axis='both', which='minor', labelsize=7, labelcolor='gray')

#绘制网格
ax.grid(True, which='major', linestyle='--', linewidth=0.7, alpha=0.7)

#绘制曲线
np.random.seed(42)
x = np.linspace(0.5, 3.5, 100)
y1 = 0.8 * np.sin(0.5 * np.pi * x) + 3
y2 = 0.5 * np.cos(0.5 * np.pi * (x - 0.2)) + 1

line1, = ax.plot(x, y1, color='#3e8abe', linewidth=2, label='Blue signal')
line2, = ax.plot(x, y2, color='#ff8d28', linewidth=2, label='Orange signal')

#绘制散点
np.random.seed(42)
scatter_x = np.random.rand(30) * 3 + 0.5
scatter_y = np.random.rand(30) * 3 + 0.5
#scatter_x = np.linspace(0.5, 3.5,20)
#scatter_y = 0.5 * np.sin(2 * np.pi * scatter_x) + 0.5 + np.random.normal(0, 0.02, 20) + 1.5
ax.scatter(scatter_x, scatter_y, color='#9467bd', marker='s')

#绘制图列
ax.legend()
#或：用函数get_legend_handles_labels()返回Axes上标识，这些标识可用于为生成的图例创建条目
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
```

## work_5：比较pyplot-style与OO-style

#### 1、风格定义

pyplot-style：是Matplotlib中的一种过程式绘图风格，通过matplotlib.pyplot模块提供的函数来进行绘图，即用户可以通过plt命令直接绘制图像，并隐式管理图形和坐标轴。

OO-style（面向对象风格）：是Matplotlib基于API接口实现的对象式绘图方式，通过显式创建图形和坐标轴（Axes/Axis）对象，然后调用这些对象的方法来进行绘图操作。

#### 2、区别

（1）、代码风格：pyplot-style是过程式编程，隐式状态管理，通过函数式调用pyplot进行绘图，代码简洁。OO-style是面向对象的显式编程，通过调用对象方法绘图，代码结构更加清晰明确。

（2）、可控性：pyplot-style依赖全局状态，通过plt.subplot()进行切换控制，多子图复杂绘制时易混淆。OO-style可以直接操作特定的Figure/Axes对象，复杂场景中不易冲突，实现灵活、精细的控制。

（3）、适用场景：pyplot-style适用简单的交互式绘图或单子图绘制，简单轻量。OO-style适用于复杂图形和多图布局GUI嵌入等的批量图形绘制，灵活便捷。

# work_5：ASCII、GB2312、Unicode编码方式应用

### 1、UTF-8编码字符集（UFT-8与Uincode间的转换）

```python
original_text = "新年快乐！"
print(f"1. 原始字符串: {original_text}")
print(f"   类型: {type(original_text)}")

utf8_bytes = original_text.encode('utf-8')
print(f"\n2. UTF-8编码字节:")
print(f"   字节对象: {utf8_bytes}")
print(f"   十六进制: {utf8_bytes.hex()}")
print(f"   每个字节: {list(utf8_bytes)}")

print(f"3. 每个字符的Unicode码点:")
for i, char in enumerate(original_text):
    code_point = ord(char)  # 获取Unicode码点（十进制）
    print(f"   '{char}' -> Unicode: U+{code_point:04X} ({code_point})")
```

### 2、字符集在ASCII与Unicode间的转换

```python
#定义字符串
text = "Hello world!"
print(f"1.原编码方式下字节类型为: {type(text)}")

#Unicode转ASCII（Python中默认字符为Unicode）
ascii_bytes = text.encode('ascii')
print(f"\n2.ASCII编码：{ascii_bytes}")
print(f"转换成ASCII后字节类型为: {type(ascii_bytes)}")

print('编成的ASCII码显示：')
for byte in ascii_bytes:
    print(f"  字符: {chr(byte)} -> ASCII值: {byte}")

#ASCII转Unicode
unicode_text = ascii_bytes.decode('ascii')
print(f"\n3.ASCII解码为Unicode：{unicode_text}")
print(f"从ASCII转换为Unicode后字节类型为: {type(unicode_text)}")
```

### 3、字符集在Uncode编码与GB2312编码间的转换

```python
#Unicode转GB2312
char_text = "新年快乐！"

gb2312_bytes = char_text.encode('gb2312')
print(f"1.GB2312编码（十六进制表示）: {gb2312_bytes.hex()}")
print(f"\n2.GB2312编码码值: {gb2312_bytes}")

#GB2312转Unicode
decoded_text = gb2312_bytes.decode('gb2312')
print(f"\n3.解码为Unicode: {decoded_text}")
```
